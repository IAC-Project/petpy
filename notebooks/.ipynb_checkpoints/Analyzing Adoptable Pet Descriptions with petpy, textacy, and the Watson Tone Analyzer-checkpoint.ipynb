# Tone Analysis of Adoptable Cat and Dog Descriptions from Petfinder

Most animals listed on Petfinder are also given a description by the shelter that provides further details and information on the pet. These descriptions are useful for increasing interest among potential adopters by helping to establish a more personal connection to the animal beyond just cute pictures (not to say I can't get enough of cute cat pictures). 

Do these descriptions vary in tone depending on the type of animal or the animal's age or other statistics? Through the combination of several Python libraries [`petpy`](http://petpy.readthedocs.io/en/latest/), [`textacy`](https://textacy.readthedocs.io/en/stable/index.html), and the [IBM Watson Tone Analyzer API](https://www.ibm.com/watson/services/tone-analyzer/), we can analyze the perceived tone of the descriptions of adoptable cats and dogs on Petfinder to see if there are any significant differences in how shelters describe animals with different ages, genders, etc.

Other Python libraries that will be used in this analysis include [`pandas`](https://pandas.pydata.org/) for working with the extracted and transformed data, [`seaborn`](https://seaborn.pydata.org/) and [`matplotlib`](https://matplotlib.org/) for data visualization, and the `stats` module available in [`scipy`](https://docs.scipy.org/doc/scipy/reference/stats.html) for hypothesis testing.

## Getting Started

Before diving in, import the libraries that we will use throughout the analysis. [`%matplotlib inline`](http://ipython.readthedocs.io/en/stable/interactive/plotting.html) is used for displaying the visualizations.

````python
%matplotlib inline

from petpy import Petfinder
import os
import textacy
import numpy as np
import pandas as pd
from watson_developer_cloud import ToneAnalyzerV3
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

import warnings
warnings.simplefilter('ignore')
````

### Authenticating the API Connections

Once the respective API credentials are obtained, we can continue to authenticate our connection to the APIs. It is recommended to obfuscate the API credentials to avoid exposing the keys to the public. One approach to doing securing the generated API keys is to store the keys as environment variables and load them using [`os.environ`](https://docs.python.org/2/library/os.html). 

````python
pf_key = os.environ['PETFINDER_KEY']
watson_tone_user = os.environ['WATSON_TONE_SERVICE_USERNAME']
watson_tone_password = os.environ['WATSON_TONE_SERVICE_PASSWORD']
````

Authenticating with the Petfinder API involves initializing the `Petfinder` class with the received key from Petfinder. The connection to the IBM Watson Tone Analyzer API is made through initializing the `ToneAnalyzerV3` class using the username and password of the service instance created earlier. Another example of authenticating with the Tone Analyzer service (and other IBM services) can be found in the [Python-SDK library on Github](https://github.com/watson-developer-cloud/python-sdk/blob/master/examples/tone_analyzer_v3.py).

````python
pf = Petfinder(pf_key)

tone_analyzer = ToneAnalyzerV3(
    username=watson_tone_user,
    password=watson_tone_password,
    version='2017-09-21')
````

Now that our connections to the needed API have been authenticated, we can begin extracting the needed data and analyzing the tones of adoptable pet descriptions!

## Obtaining and Preparing the Data

Our strategy for preparing the pet adoption data is as follows:

* Extract a sample of adoptable cat and dog descriptions from the Petfinder API.
* Combine the extracted data and remove unnecessary data points.
* Preprocess the provided animal descriptions to clean the text.
* Obtain the tone scores of the cleaned pet descriptions using the IBM Watson Tone Analyzer.

### Extracting Adoptable Cat and Dog Descriptions from Petfinder

As we only have 2,500 free calls to the Tone Analyzer API, we set the amount of results to return from the Petfinder database to 1,200 cat and dog records. The [`pet_find`](http://petpy.readthedocs.io/en/latest/api.html#pet-methods) method makes it easy to extract a sample of adoptable pet information from the database.

````python
cats = pf.pet_find(location='United States', count=600, pages=2, animal='cat', return_df=True)
dogs = pf.pet_find(location='United States', count=600, pages=2, animal='dog', return_df=True)
````

The cat and dog data extracted from the Petfinder database are then concatenated together using [`pd.concat`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.concat.html). We also print the shape of the `DataFrame` to ensure the number of records are what we would expect. As we are extracting 1,200 cat and dog records, the total number of rows should be 2,400.

````python
cat_dog = pd.concat([cats, dogs])
cat_dog.shape
````

To avoid having to call the Petfinder API more than necessary and for reproducible results, it is a good step to save the raw extracted results from the API into a CSV or other format before applying any transformations and data cleaning.

````python
cat_dog.to_csv('../data/cat_dog.csv', index=False, encoding='utf-8')
````

We also want to make sure all the pet records have a description available for analysis. Therefore, we filter any records that have a missing description before proceeding. After removing empty descriptions, we print the new shape of the `DataFrame` to see how many records were removed.

````python
cat_dog = cat_dog.loc[pd.isnull(cat_dog['description']) == False]
cat_dog.shape
````

Let's also inspect the columns in the current `DataFrame` using the `columns` attribute.

````python
cat_dog.columns
````

The `photo` columns are removed as they are unneeded for this analysis as well as a few other miscellaneous columns such as the shelter-related columns.

````python
cat_dog = cat_dog.loc[:, list(cat_dog.columns[0:list(cat_dog.columns).index('phone')]) + ['sex', 'size', 'status']]
````

### Preprocessing the Pet Descriptions

Text preprocessing is an important step when analyzing text such as adoptable pet descriptions. Fortunately, the Petfinder platform provides something of a standardized way to create pet descriptions, therefore, the text we will be dealing with will likely be more clean than other sources (social media and forums, for example). Also, we want to keep the descriptions unaltered as much as possible in order to preserve the original tone and quality of the description. 

[`textacy`](https://textacy.readthedocs.io/en/stable/index.html) is a wonderful library for processing and analyzing text that is built on top of the natural language library [`spaCy`](https://spacy.io/). The textacy library makes it much more straightforward to preprocess text with its [`preprocessing`](https://textacy.readthedocs.io/en/stable/api_reference.html#module-textacy.preprocess) module. There are many text processing methods available in [`textacy`]; however, we will only take advantage of the `fix_bad_unicode` and `normalize_whitespace` methods to hopefully normalize the pet descriptions to get better results from the Watson Tone Analyzer. 

The preprocessing methods in the `preprocess` module are used in pandas' [`apply`](http://pandas.pydata.org/pandas-docs/version/0.18/generated/pandas.Series.apply.html) method. A new column `description_clean` is created to store the new descriptions to perserve the original descriptions just in case.

````python
cat_dog['description_clean'] = cat_dog['description'].apply(
    lambda x: textacy.preprocess.fix_bad_unicode(str(x))).apply(
    lambda x: textacy.preprocess.normalize_whitespace(str(x)))
````

### Analyzing the Tone of the Cat and Dog Descriptions

We can now use the cleaned description text as input for the Watson Tone Analyzer API. For each row in the `DataFrame`, we run the `tone` method of the `ToneAnalyzerV3` class we initialized and set the `content_type` parameter to plain text and the `sentences` parameter to `False` as we are not interested in sentence level tone analysis for this task. We iterate through the `DataFrame` rows using the [`iterrows()`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iterrows.html) method.

As with most APIs, the return data type from the Tone Analyzer API is a JSON object. Therefore, we must coerce the result in a tabular structure to make it easier for analysis. Fortunately, `pandas` provides a very handy function, [`json_normalize`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.json.json_normalize.html), for normalizing structured JSON into a flat data table. Thus, we will leverage this function to coerce the Tone Analyzer JSON results into a pandas `DataFrame` and combine the respective pet's data into a new `DataFrame`.  

````python
cat_dog_tones = pd.DataFrame()

for idx, row in cat_dog.iterrows():
    tones = tone_analyzer.tone(row['description_clean'], content_type='text/plain;charset=utf-8', sentences=False)
    tones_df = pd.io.json.json_normalize(tones['document_tone'], 'tones')
    
    tones_df[cat_dog.columns.tolist()] = pd.DataFrame([row], index=tones_df.index)
    cat_dog_tones = cat_dog_tones.append(tones_df)
````

Once the iteration through the `DataFrame` is complete and we have combined the output Watson Tone Analyzer API with the original data, we inspect the first few rows of the resulting `DataFrame` to make sure the output is what we expect.

We can see the Tone Analyzer API data `tone_id`, `tone_name` and `score` have been appended to the data! As before, it is often a good idea to save the results to a file to avoid having to call any APIs more than necessary (and incurring any additional fees, etc.).

````python
cat_dog_tones.to_csv('../data/cat_dog_tones.csv', index=False, encoding='utf-8')
````

We have accomplished the first step of extracting and preparing the data for analysis! We now proceed to investigate the scores of the tones in the adoptable pet descriptions as analyzed by the algorithms used under the hood by the Watson Tone Analyzer API. Using the scores output from the Tone Analyzer API will allow us to see if there are any significant differences in how shelters represent adoptable cats and dogs.

## Analysis

Before beginning, we set the base style of the `seaborn` plots using the [`set`](https://seaborn.pydata.org/generated/seaborn.set.html) function.

````python
sns.set(style="white", context="talk", font_scale=1.5)
````

````python
cat_dog_pivot = pd.pivot_table(cat_dog_tones, 
                               index=['animal', 'age', 'sex', cat_dog_tones.index], 
                               columns=['tone_name'], values=['score'], 
                               aggfunc=np.mean).reset_index()
````

````python
cat_descrip = pd.DataFrame({'count': cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score'].count(), 
             'percentage': cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score'].count() / cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat'].shape[0] * 100,
                           'animal': 'Cat'})

dog_descrip = pd.DataFrame({'count': cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score'].count(), 
             'percentage': cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score'].count() / cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog'].shape[0] * 100,
                           'animal': 'Dog'})
````

````python
tone_counts = pd.concat([cat_descrip, dog_descrip]).reset_index()
tone_counts
````

````python
plt.figure(figsize=(10, 8))

b = sns.factorplot(x='tone_name', y='percentage', hue='animal', kind='bar', data=tone_counts, size=10, legend=False)
b.set(xlabel='Tones', ylabel='Percentage of Descriptions with Tone')

sns.despine(bottom=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
f, ax = plt.subplots(2, 1, figsize=(18, 14), sharex=True)

k = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score']['Analytical'], 
                label='Analytical', shade=True, ax=ax[0])
k = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score']['Confident'], 
                label='Confident', shade=True, ax=ax[0])
k = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score']['Joy'], 
                label='Joy', shade=True, ax=ax[0])
k = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score']['Sadness'], 
                label='Sadness', shade=True, ax=ax[0])
k = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Cat']['score']['Tentative'], 
                label='Tentative', shade=True, ax=ax[0])

k.set(xlim=(0.5,1), title='Distribution of Tone Scores of Adoptable Cat Descriptions', 
      ylabel='Density', xlabel='Tone Score')

k1 = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score']['Analytical'], 
                 label='Analytical', shade=True, ax=ax[1], legend=False)
k1 = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score']['Confident'], 
                 label='Confident', shade=True, ax=ax[1], legend=False)
k1 = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score']['Joy'], 
                 label='Joy', shade=True, ax=ax[1], legend=False)
k1 = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score']['Sadness'], 
                 label='Sadness', shade=True, ax=ax[1], legend=False)
k1 = sns.kdeplot(cat_dog_pivot[cat_dog_pivot['animal'] == 'Dog']['score']['Tentative'], 
                 label='Tentative', shade=True, ax=ax[1], legend=False)

k1.set(xlim=(0.5,1), title='Distribution of Tone Scores of Adoptable Dog Descriptions', 
       ylabel='Density', xlabel='Tone Score')

sns.despine(bottom=True)

k.legend(bbox_to_anchor=(1.05, 0), loc=2, borderaxespad=0.)
````

````python
mean_tone_scores = pd.pivot_table(cat_dog_tones, 
                                  index=['animal', 'age', 'sex'], 
                                  columns=['tone_name'], values=['score'], 
                                  aggfunc=np.mean).reset_index()
````

````python
mean_tone_scores = mean_tone_scores.melt(id_vars=['animal', 'age', 'sex'], value_vars='score')
````

````python
mean_tone_scores = mean_tone_scores.loc[(mean_tone_scores['tone_name'] != 'Fear') & 
                    (mean_tone_scores['tone_name'] != 'Anger')]
````

````python
g = sns.factorplot(data=mean_tone_scores, x='tone_name', y='value', hue='sex', col='animal', kind='bar', size=10, col_wrap=2, ci=None)
g.set(ylim=(0.5,1), xlabel='Tone', ylabel='Score')
sns.despine(bottom=True)
````

````python
g = sns.factorplot(data=mean_tone_scores, x='tone_name', y='value', 
                   hue='age', col='animal', kind='bar', size=12, col_wrap=2, ci=None, sharey=True)
g.set(ylim=(0.5,1), xlabel='Tone', ylabel='Score')
sns.despine(bottom=True)
````
